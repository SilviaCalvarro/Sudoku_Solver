## 2. Algoritmo Backtracking 

```python
#Función de printeo del grid
def printGrid():
    global sudoky_matrix
    fig = plt.figure(figsize = (10,10))
    sns.heatmap(sudoky_matrix, annot=True, cbar=None, xticklabels=False, yticklabels=False)

    plt.show()
```

```python
#Función para definir las restricciones de las celdas
def possible(row,col,digit):
    global sudoky_matrix
    for i in range(0,9):
        if sudoky_matrix[row][i] == digit:
            return False
    for i in range(0,9):
        if sudoky_matrix[i][col] == digit:
            return False
    square_row = (row//3)*3
    square_col = (col//3)*3
    for i in range(0,3):
        for j in range(0,3):
            if sudoky_matrix[square_row+i][square_col+j] == digit:
                return False    
    return True
```

```python
def solve():
    global sudoky_matrix
    for row in range(9):
        for col in range(9):
            if sudoky_matrix[row][col] == 0:
                for digit in range(1,10):
                    if possible(row,col,digit):
                        sudoky_matrix[row][col] = digit
                        solve()
                        sudoky_matrix[row][col] = 0 
                return
    print('\nLa solución es:')
    printGrid()
```

## 3. Solución

```python
solve()
```
